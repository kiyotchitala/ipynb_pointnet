normalize.txtの点群を見るノートブックです


```python
import glob
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

files = glob.glob('./normalize_txt/takuya/150cm/*')
if len(files) == 0:
    print("No data available")
else:
    randnum = random.randint(0,len(files))
    with open(files[randnum],'r') as f:
        datas = f.read().split('\n')

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection='3d')
for data in datas:
    if not data == '':
        d_s = data.split(' ')
        ax.scatter(float(d_s[0]),float(d_s[1]),float(d_s[2]),s=40,c='blue')
plt.show()
```
